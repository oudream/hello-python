```python

arr = [1, 2, 4, 1, 7, 8, 3]

def rec_opt(arr, k):
    if (k == 0):
        return arr[0]
    elif (k == 1):
        return max(arr[0], arr[1])
    else:
        r1=rec_opt(arr, k-1)
        r2=rec_opt(arr, k-2)+arr[k]
        return max(r1, r2)
    
def loop_opt(arr):
    import numpy as np
    opt = np.zeros(len(arr))
    opt[0] = arr[0]
    opt[1] = max(arr[0], arr[1])
    for i in range(2, len(arr)):
        r1 = opt[i-1]
        r2 = opt[i-2] + arr[i]
        opt[i] = max(r1, r2)
    return opt[len(arr)-1]
    
print(rec_opt(arr,6))
print(loop_opt(arr))
```
